```python
import numpy as np
houses = np.array([
    [1200, 3, 200000],
    [1500, 4, 250000],
    [1000, 2, 150000],
    [1800, 4, 300000]
])

print("Dataset:")
print(houses)
```

```python
x = houses[:, :-1]
print("Features (X):")
print(x)
```

```python
y = houses[:, -1]
print("Target (y):")
print(y)
```
