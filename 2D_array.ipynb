```python
import numpy as np
data = np.array([
    [25, 2000],
    [30, 3000],
    [35, 4000]
])

print(data)
```

```python
print("shape:", data.shape)
```

```python
print(data[0, 0]) # age of first row
print(data[0, 1]) # Salary of first row
```

```python
#all rows,  column 0
ages = data[:, 0]

#all rows, column 1
salaries = data[:, 1]

print("ages:", ages)
print("salaries:", salaries)
```
